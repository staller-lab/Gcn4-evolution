```python
def read_fasta_file(filename):
    """
    Reads a FASTA file and returns a list of protein sequences.
    :param filename: name of the FASTA file to read
    :return: a list of protein sequences as strings
    """
    sequences = []
    with open(filename) as f:
        sequence = ""
        for line in f:
            line = line.strip()
            if line.startswith(">"):
                # Found a new sequence header, so add the previous sequence to the list and start a new one
                if sequence:
                    sequences.append(sequence)
                    sequence = ""
            else:
                # Add this line to the current sequence
                sequence += line
        # Add the last sequence to the list
        if sequence:
            sequences.append(sequence)
    return sequences
```

```python
import numpy as np

def calculate_percentage_identity(sequences):
    num_sequences = len(sequences)
    sequence_length = len(sequences[0])
    identity_array = np.zeros(sequence_length)

    for i in range(sequence_length):
        residues_at_position = [sequence[i] for sequence in sequences if sequence[i] != '-']
        if len(residues_at_position) == 0:
            identity_array[i] = 0
        else:
            most_common_residue = max(set(residues_at_position), key=residues_at_position.count)
            num_common_residues = residues_at_position.count(most_common_residue)
            percentage_identity = (num_common_residues / num_sequences) * 100
            identity_array[i] = percentage_identity

    return identity_array
```

```python
import numpy as np
import pandas as pd
from Bio.Align import substitution_matrices
from scipy import stats

#what rate do they change from one to another
def compute_position_score(msa, position, matrix=substitution_matrices.load("BLOSUM62")):
    column = [seq[position] for seq in msa]
    pairs = [(a, b) for idx, a in enumerate(column) for b in column[idx + 1:] if a != '-' and b != '-']
    scores = [matrix[pair] if pair in matrix else matrix[pair[::-1]] for pair in pairs]
    return np.mean(scores) if scores else np.nan

def calculate_region_score(msa, start, end):
    return np.nanmean([compute_position_score(msa, position) for position in range(start, end)])

def compare_regions(msa, abd1_start, abd1_end, entire_start, entire_end):
    abd1_score = calculate_region_score(msa, abd1_start, abd1_end)
    entire_score = calculate_region_score(msa, entire_start, entire_end)
    return abd1_score, entire_score

def perform_statistical_test(msa, abd1_start, abd1_end, entire_start, entire_end):
    abd1_scores = [compute_position_score(msa, position) for position in range(abd1_start, abd1_end)]
    entire_scores = [compute_position_score(msa, position) for position in range(entire_start, entire_end)]
    abd1_scores_clean = [score for score in abd1_scores if not np.isnan(score)]
    entire_scores_clean = [score for score in entire_scores if not np.isnan(score)]
    u_stat, p_value = stats.mannwhitneyu(abd1_scores_clean, entire_scores_clean, alternative='two-sided')
    return u_stat, p_value
```

```python
msa = sequences
abd1_start, abd1_end = 158, 238
entire_start, entire_end = 0, len(msa[0])  # Assuming all sequences in MSA have the same length
abd1_score, entire_score = compare_regions(msa, abd1_start, abd1_end, entire_start, entire_end)
u_stat, p_value = perform_statistical_test(msa, abd1_start, abd1_end, entire_start, entire_end)

print("ABD1 average conservation score:", abd1_score)
print("Entire protein average conservation score:", entire_score)
print("U-statistic:", u_stat)
print("P-value:", p_value)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_percent_identity(amino_acid_positions, percent_identity, regions_of_interest=None, filename=None):
    plt.figure(figsize=(10, 6))
    plt.plot(amino_acid_positions, percent_identity, label='Percent Identity', color='blue')
    plt.fill_between(amino_acid_positions, percent_identity, color='blue', alpha=0.1)

    # Adding regions of interest
    if regions_of_interest:
        for region, label in regions_of_interest.items():
            start, end = region
            plt.axvspan(start, end, color='red', alpha=0.2)  # shade region
            plt.text((start + end) / 2, 104, label,
                     horizontalalignment='center', verticalalignment='top', color='red')

    plt.title('Percent Identity of Gal 11 ABD1 Compared to YGOB')
    plt.xlabel('Amino Acid Position')
    plt.ylabel('Percent Identity')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    if filename:
        plt.savefig(filename, bbox_inches='tight')
    else:
        plt.show()
```

```python
# First, we'll modify the existing function to include the ability to mark specified residues with asterisks.
# These residues are the ones that are known to interact with GCN4.

def plot_percent_identity_with_touching_residues(amino_acid_positions, percent_identity, regions_of_interest=None, touching_residues=None, filename=None):
    plt.figure(figsize=(10, 6))
    plt.plot(amino_acid_positions, percent_identity, label='Percent Identity', color='blue')
    plt.fill_between(amino_acid_positions, percent_identity, color='blue', alpha=0.1)

    # Adding regions of interest
    if regions_of_interest:
        for region, label in regions_of_interest.items():
            start, end = region
            plt.axvspan(start, end, color='red', alpha=0.2)  # shade region
            plt.text((start + end) / 2, 104, label,
                     horizontalalignment='center', verticalalignment='top', color='red')

    # Marking the residues that touch GCN4 with an asterisk
    if touching_residues:
        for residue in touching_residues:
            index = residue - amino_acid_positions[0]  # Adjusting for the start position
            if 0 <= index < len(amino_acid_positions):  # Check if within the range
                plt.plot(amino_acid_positions[index], percent_identity[index], 'r*', markersize=10)

    plt.title('Percent Identity of Gal 11 ABD1 Compared to YGOB')
    plt.xlabel('Amino Acid Position')
    plt.ylabel('Percent Identity')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    if filename:
        plt.savefig(filename, bbox_inches='tight')
    else:
        plt.show()

# Now we'll call this updated function with the additional parameter for touching residues.
# The positions must be zero-indexed for proper alignment on the plot, so we subtract 1 from each.
```
