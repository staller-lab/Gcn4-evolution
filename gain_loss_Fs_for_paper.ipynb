```python
def read_fasta(file):
    seqs = []
    names = []
    seq = ""

    with open(file, "r") as f:
        lines = f.readlines()
    
    for line in lines:
        if line.startswith(">"):
            names.append(line.strip().replace(">",""))
            if len(names) > 1:
                seqs.append(seq)
                seq = ""
        else:
            seq += line.strip()

    seqs.append(seq)
    return pd.DataFrame({"name" : names, "aa_seq" : seqs})
```

```python
def get_coords(alignment, name, seq_df):

    alignment_df = read_fasta(alignment)
    subset_alignment_df = alignment_df[alignment_df['name'] == name ]

    subset_seq_df = seq_df[seq_df['name'] == name]
    
    start, end = subset_seq_df["AD_coords"].values[0][0]
    ad_seq = subset_seq_df['sequence'].values[0][start:end]
    print(ad_seq)
    i = 0
    start = -1
    for j, aa in enumerate(subset_alignment_df['aa_seq'].values[0]):
        if aa == ad_seq[i]:
            if start == -1:
                start = j
            i += 1
        elif aa == "-":
            continue
        else:
            i = 0
            start = -1
        
        if i == len(ad_seq):
            end = j
            return start, end
```
