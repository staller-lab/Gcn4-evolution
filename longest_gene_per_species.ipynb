```python
import pandas as pd
```

```python
# Load a dataframe mapping sequence names to their species
name_key = pd.read_csv("../data/phylogenetic_info/species_mapping.csv")
name_key
```

```python
# Read in the experimental data (full length sequences and smoothed activity data)
FullOrthoDF =  pd.read_pickle('../data/FullOrthologDF_20240930')
FullOrthoDF
```

```python
# Merge with the species name df
FullOrthoDF = FullOrthoDF.rename(columns={"SpeciesName" : "HomologName"})
FullOrthoDF = FullOrthoDF.merge(name_key, on = "HomologName")
```

```python
# Sort by longest and shortest
longest = FullOrthoDF.sort_values("Length", ascending=False).drop_duplicates("SpeciesName")
shortest = FullOrthoDF.sort_values("Length", ascending=True).drop_duplicates("SpeciesName")
```

```python
# Write longest to fasta file
with open("../data/fasta_files/longest_gene_per_species.fasta", "w+") as f:
    for i in longest.index:
        f.write(">" + longest.loc[i, "HomologName"] + "\n")
        f.write(longest.loc[i, "Seq"] + "\n")
```

```python
# Save df of longest to pickle
longest.to_pickle('../data/FullOrthologDF_longest_gene_per_species_20240930')
```
