```python
import pandas as pd
from ete3 import TreeStyle, PhyloTree, Tree, faces, NodeStyle, AttrFace, TextFace
from ete3.treeview.faces import add_face_to_node
```

```python
t = PhyloTree("../data/phylogenetic_info/all_acromyces.newik", format=1)
```

```python
name_key = pd.read_csv("../data/phylogenetic_info/top138_phyloinfo.csv")
leaves = set(list(name_key["TreeLeaf"].dropna()) + ["Lacth1", "Sacce1", "Cantro1"])
prune_leaves = [leaf for leaf in leaves if leaf in t.get_leaf_names()]
t.prune(prune_leaves)
```
